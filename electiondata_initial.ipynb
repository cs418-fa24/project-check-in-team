```python

df = pd.read_csv('/Users/purvatandel/Desktop/modified_zillow_data.csv')

df_melted = df.melt(id_vars=['StateName'], var_name='Date', value_name='HousePrice')

df_melted['Date'] = pd.to_datetime(df_melted['Date'])


df_melted['Year'] = df_melted['Date'].dt.year

df_yearly_avg = df_melted.groupby(['Year', 'StateName'])['HousePrice'].mean().reset_index()

election_results = {
    2000: 'Republican',
    2004: 'Republican',
    2008: 'Democrat',
    2012: 'Democrat',
    2016: 'Republican',
    2020: 'Democrat',
    2024: 'Republican',  
}

df_yearly_avg['WinningParty'] = df_yearly_avg['Year'].map(election_results)

df_pivoted = df_yearly_avg.pivot(index='Year', columns='StateName', values='HousePrice')

df_pivoted.loc['WinningParty'] = df_yearly_avg.groupby('Year')['WinningParty'].first()

df_pivoted.to_csv('/Users/purvatandel/Desktop/yearly_avg_house_prices_with_party.csv')

print(df_pivoted)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/Users/purvatandel/Desktop/yearly_avg_house_prices_with_party.csv'
data = pd.read_csv(file_path)

sns.set(style="white")  

plt.figure(figsize=(12, 6))

for state in data.columns[2:]:  
    plt.plot(data['Year'], data[state], label=state, alpha=0.7)

for idx, row in data.iterrows():
    plt.axvline(x=row['Year'], color='gray', linestyle='--', linewidth=1)
    plt.text(row['Year'], max(data.iloc[:, 2:].max()) + 10000, row['Winning Party'], 
             horizontalalignment='center', fontsize=12, color='black')

election_years = [2000, 2004, 2008, 2012, 2016, 2020, 2024]
plt.xticks(election_years)

plt.xlabel('Election Year', fontsize=14)
plt.ylabel('Average House Price', fontsize=14)
plt.title('Average House Prices by State and Election Years (2000 - 2024)', fontsize=16)

plt.xticks(rotation=45)

plt.legend(title='States', bbox_to_anchor=(1.05, 1), loc='upper left')
#plt.yscale('log')
plt.tight_layout()
plt.show()
```
