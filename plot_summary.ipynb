```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
predicted_df = pd.read_csv("./predicted_values.csv")
time_range = ["1","5","10","all"]
```

```python
summary=[]
labels = ["One Month Later", "One Quarter Later", "One year Later"]
iter = 0
for times in time_range:
   spec =  predicted_df[predicted_df["time_used"]==times]
   #print(f"TIME RANGE:{times}")
   summary.append(spec.iloc[:,2:-1].mean())
   #print(spec.iloc[:,1:-1].mean())

df = pd.DataFrame(summary)
print(df)
c =0
for i in df.iloc[:,:3].columns:
   plt.plot(time_range, df[i], marker='o', label=labels[iter])
   plt.legend(loc='upper right')     #plt.show()
   plt.xlabel("Previous Years Considered")
   plt.ylabel("Price Change Direction Accuracy")
   iter+=1
```

```python
iter = 0
for i in df.iloc[:,3:].columns:
   plt.plot(time_range, df[i], marker='o', label=labels[iter])
   plt.legend(loc='upper right')
   plt.xlabel("Previous Years Considered")
   plt.ylabel("")
   iter+=1
```
