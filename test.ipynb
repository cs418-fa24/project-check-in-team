## Descriptive Analysis of the Data

Summurizing the data and understanding the distribution of the data is very important before we start building the model. This will help us to understand the data better and also to identify the outliers in the data.

### Data Description

Measure of central tendency

- Median
- Mode

Measure of dispersion
- Range
- Variance
- Standard Deviation

```python
# Convert date columns to datetime if needed and sort by date
date_columns = df1.columns[9:] # the first 9 columns are metadata
df1[date_columns] = df1[date_columns].apply(pd.to_datetime, errors='coerce')

# Handle missing values - for example, forward fill NaN values
df1.fillna(method='ffill', inplace=True)

# Convert datetime columns to numeric values (e.g., Unix timestamp) for pct_change calculation
df1[date_columns] = df1[date_columns].apply(lambda x: x.view(np.int64))

# Calculate monthly percent change for property values for each ZIP code
df1_pct_change = df1.groupby('RegionName')[date_columns].pct_change(axis=1)

# Adding zipcode column to the df1_pct_change
df1_pct_change.insert(0, 'RegionName', df1['RegionName'])

# Display the trend over time for each ZIP code
print(df1_pct_change)

# Save the results to a CSV file
df1_pct_change.to_csv("df1_pct_change.csv", index=False)
```

```python
# Convert date columns to datetime if needed and sort by date
date_columns = df2.columns[9:] # the first 9 columns are metadata
df2[date_columns] = df2[date_columns].apply(pd.to_datetime, errors='coerce')

# Handle missing values - for example, forward fill NaN values
df2.fillna(method='ffill', inplace=True)
df2.fillna(method='bfill', inplace=True)

# Convert datetime columns to numeric values (e.g., Unix timestamp) for pct_change calculation
df2[date_columns] = df2[date_columns].apply(lambda x: x.view(np.int64))

# Calculate monthly percent change for property values for each ZIP code
df2_pct_change = df2.groupby('State')[date_columns].pct_change(axis=1)

# Adding zipcode column to the df1_pct_change
df2_pct_change.insert(0, 'State', df2['State'])

# Display the trend over time for each ZIP code
print(df2_pct_change)

# Save the results to a CSV file
df2_pct_change.to_csv("df2_pct_change.csv", index=False)
```

```python
df8 = pd.read_csv("df2_pct_change.csv")

# Select only numeric columns
numeric_columns = df8.select_dtypes(include=[np.number]).columns

# Group by State and calculate the median for each date column
df_median = df8.groupby("State")[numeric_columns].median()

df_median.to_csv("df_state_median_pct_change.csv", index=False)
```

```python
import plotly.express as px

# Sample data for all states across multiple years for each month
data = []
for state in df_median.index:
    for date in df_median.columns:
        year, month = date.split('-')[:2]  # Split date into year and month
        data.append({
            "State": state,
            "Code": state,
            "Year": int(year),
            "Month": int(month),
            "DataValue": df_median.loc[state, date]
        })

# Convert to DataFrame
df = pd.DataFrame(data)

# Create a new column for YearMonth
df['YearMonth'] = df['Year'].astype(str) + '-' + df['Month'].astype(str)

# Create an interactive choropleth map with the Viridis color scale
fig = px.choropleth(df,
                    locations="Code",
                    locationmode="USA-states",
                    color="DataValue",
                    hover_name="State",
                    animation_frame="YearMonth",
                    animation_group="State",
                    color_continuous_scale=px.colors.sequential.Viridis,
                    range_color=(-0.02, 0.02),
                    scope="usa",
                    title="U.S. State Data Showing Median Percentage Change in Property Values Across Years and Months")

# Show the map
fig.show()
```

```python
# Convert YearMonth to datetime format and sort the dataframe
df['YearMonth'] = pd.to_datetime(df['YearMonth'])
df = df.sort_values('YearMonth')

# Plotting a line plot showing the median percentage change in property values for each state over time
fig = px.line(df, x='YearMonth', y='DataValue', color='State', title='Median Percentage Change in Property Values by State Over Time')

# Update y-axis range
fig.update_yaxes(range=[-0.05, 0.05])

# Show the plot
fig.show()

# Merging all the state lines to get a smoothed line
df_smooth = df.groupby('YearMonth')['DataValue'].mean().reset_index()

# Plotting a line plot showing the average of median percentage change in property values over all states over time
fig = px.line(df_smooth, x='YearMonth', y='DataValue', title='Avg of Median Percentage Change in Property Values Over All States Over Time')

# Update y-axis range
fig.update_yaxes(range=[-0.05, 0.05])

# Show the plot
fig.show()
```

```python
# Flipping the previous plot across the x-axis, since mortgage rates are inversely related to property values

# Plotting a line plot
fig = px.line(mort_df, x='DATE', y='MORTGAGE30US', title='24-Year Fixed Mortgage Rate Over Time Reversed')

# Update y-axis range
fig.update_yaxes(autorange="reversed")

# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go

# Plotting this reversed mortgage plot along with the smoothed line plot of the median percentage change in property values for all states
fig = px.line(df_smooth, x='YearMonth', y='DataValue', title='Avg of Median Percentage Change in Property Values Over All States Over Time')

# Update y-axis range
fig.update_yaxes(range=[-0.015, 0.025])

# Add a secondary y-axis with reversed mortgage rates
fig.add_trace(
    go.Scatter(x=mort_df['DATE'], y=mort_df['MORTGAGE30US'], mode='lines', name='24-Year Fixed Mortgage Rate Reversed', yaxis='y2')
)

# Update layout with proper labels and grid lines
fig.update_layout(
    yaxis=dict(title='Median Percentage Change in Property Values', showgrid=True),
    yaxis2=dict(title='24-Year Fixed Mortgage Rate Reversed', overlaying='y', side='right', autorange='reversed', showgrid=False),
    xaxis=dict(title='YearMonth', showgrid=True),
    title=dict(text='Avg of Median Percentage Change in Property Values Over All States Over Time', x=0.5),
    legend=dict(x=0.01, y=0.99),
    template='plotly_white'
)

# Show the plot
fig.show()
```
